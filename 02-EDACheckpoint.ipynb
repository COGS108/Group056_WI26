### Dataset #1 

Instructions: REPLACE the contents of this cell and the one below with your work, including any updates to recover points lost in your data checkpoint feedback

This dataset contains 2,149 patient records and 35 variables related to demographics, cardiovascular health, metabolic measures, and Alzheimer’s disease diagnosis. Each row represents a single patient and includes demographic information, clinical health indicators, cognitive scores, and binary indicators for medical conditions.

Dataset Name: Alzeimer's Clinical Dataset by Vinod Vadde
Link to the dataset: https://www.kaggle.com/datasets/vinodvadde/alzeimers-clinical-dataset
Number of observations: 2,149 patient records
Number of variables: 35 variables (16 binary categorical)

Key variables:
- Age (years): Ranges from 60–90 and is the strongest known risk factor for Alzheimer’s disease, as risk increases significantly with age.
- EducationLevel (0-3): Represents the patient’s level of education, which may influence cognitive reserve and resilience to cognitive decline.
- BMI (kg/m²): Body mass index measuring body composition. Healthy values typically range from 18.5-24.9.
- SystolicBP / DiastolicBP (mmHg): Blood pressure measurements; normal levels are generally below 120/80 mmHg.
- CholesterolTotal, LDL, HDL, Triglycerides (mg/dL): Indicators of metabolic and cardiovascular health.
- MMSE (0–30): Mini-Mental State Examination score measuring cognitive function; scores above 25 are considered normal.
- ADL and FunctionalAssessment (0-10): Measure a patient’s ability to perform daily tasks independently.
- Binary health indicators (0 = No, 1 = Yes): Variables such as Diabetes, Depression, Confusion, Disorientation, and PersonalityChanges indicate the presence of specific conditions or symptoms.
- Diagnosis: Indicates whether the patient has Alzheimer’s disease and serves as the primary outcome variable.

Because our research question focuses on how demographic factors relate to Alzheimer’s diagnosis, it is important that all demographic and clinical variables are measured on the same patients. Although our hypothesis highlights age as the strongest predictor, Alzheimer’s disease is influenced by a combination of demographic and health-related factors. Clinical and metabolic variables such as blood pressure, cholesterol levels, and diabetes may interact with age and help determine whether age remains the dominant predictor after accounting for overall health.

After removing the identifier variables (PatientID and DoctorInCharge), 32 variables remain in the dataset. While all variables are explored during the initial analysis, we focus primarily on demographic and clinical variables most relevant to Alzheimer’s diagnosis.

One major concern with this dataset is self-selection bias. The people in this study are likely those who have the time and money to visit a doctor regularly or use health tracking apps. This means that the sample is probably skewed towards people with higher incomes and better education. Due to the fact that these individuals often have a brain which is, in a way, more resilient to damage, a model trained on them might not work very well for the general public, especially for people in lower-income areas who don’t have the same access to healthy food or medical care.

Another issue is that many variables, such as DietQuality or PhysicalActivity, are usually self-reported. This would lead to recall bias, where people, especially those already experiencing symptoms of memory loss, might not remember their habits accurately. People also tend to round up/down how much they exercise or how much they drink to look better on paper. If the data contains such inconsistencies resulting from the people not being completely honest or accurate, an ML model might end up chasing patterns that aren’t actually real, making its predictions less reliable in a real clinical setting.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

df = pd.read_csv("data/00-raw/alzheimers_disease_data.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns)
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
print("Rows before:", df.shape[0])

df = df.dropna()

print("Rows after:", df.shape[0])
```

```python
print("Duplicate rows:", df.duplicated().sum())
df = df.drop_duplicates()
```

```python
categorical_cols = [
    "Gender","Ethnicity","EducationLevel", "Smoking","FamilyHistoryAlzheimers",
    "CardiovascularDisease","Diabetes","Depression",
    "HeadInjury","Hypertension", "MemoryComplaints", "BehavioralProblems",
    "Confusion", "Disorientation", "PersonalityChanges", "DifficultyCompletingTasks", 
    "Forgetfulness", "Diagnosis"
]

for col in categorical_cols:
    df[col] = df[col].astype("category")
```

```python
df = df.drop(columns=["PatientID", "DoctorInCharge"])
```

```python
num = df.select_dtypes(include=np.number)

print("num shape:", num.shape)
print("unique colnames:", num.columns.nunique(), "total:", len(num.columns))

# show duplicate column names (if any)
print("duplicate colnames:", num.columns[num.columns.duplicated()].tolist()[:20])
```

```python
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_cols = df.select_dtypes(include="number").columns

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    df[col] = df[col].clip(q1 - 1.5*iqr, q3 + 1.5*iqr)
```

```python
summary_cols = [
    'Age', 'BMI', 'AlcoholConsumption', 'PhysicalActivity',
    'DietQuality', 'SleepQuality', 'SystolicBP', 'DiastolicBP',
    'CholesterolTotal', 'CholesterolLDL', 'CholesterolHDL',
    'CholesterolTriglycerides', 'MMSE', 'FunctionalAssessment', 'ADL'
]

print("Summary statistics for key health features:")
print(df[summary_cols].describe().round(2).to_string())
print()

print(f"Number of unique diagnoses: {df['Diagnosis'].nunique()}")

print("\nPatient count per diagnosis:")
print(df['Diagnosis'].value_counts().sort_index())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("data/00-raw/alzheimers_disease_data.csv")

df.head()
```

Diagnosis distribution 


```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Diagnosis', data=df, palette='viridis')
plt.title('Distribution of Alzheimer\'s Diagnosis')
plt.xlabel('Diagnosis (0: No, 1: Yes)')
plt.ylabel('Count')
plt.show()
```

Age distribution

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='Age', hue='Diagnosis', fill=True, common_norm=False, palette='magma')
plt.title('Age Distribution by Diagnosis')
plt.show()
```

Age vs MMSE scatterplot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='MMSE', hue='Diagnosis', alpha=0.5)
plt.title('MMSE Score vs Age colored by Diagnosis')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(ax=axes[0], x='Diagnosis', y='DietQuality', data=df, palette='Set2')
axes[0].set_title('Diet Quality vs Diagnosis')

sns.boxplot(ax=axes[1], x='Diagnosis', y='SleepQuality', data=df, palette='Set2')
axes[1].set_title('Sleep Quality vs Diagnosis')

plt.show()
```

Box plot for Age

```python
plt.figure(figsize=(10, 6))
sns.boxenplot(x='Diagnosis', y='Age', data=df)
plt.title("Age Distribution by Diagnosis (Hypothesis Check)")
plt.xlabel("Diagnosis (0 = Healthy, 1 = Alzheimer's)")
plt.ylabel("Age")
plt.show()
```

Spiderchart

```python
features = ['MMSE', 'FunctionalAssessment', 'ADL', 'PhysicalActivity', 'DietQuality']

scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[features]), columns=features)
df_scaled['Diagnosis'] = df['Diagnosis']

avg_profile = df_scaled.groupby('Diagnosis').mean().reset_index()

plt.figure(figsize=(10, 5))
avg_profile_melt = avg_profile.melt(id_vars='Diagnosis', var_name='Feature', value_name='Normalized Value')
sns.barplot(data=avg_profile_melt, x='Feature', y='Normalized Value', hue='Diagnosis')
plt.title("Clinical Profile: Alzheimer's vs Healthy (Normalized Features)")
plt.ylabel("Relative Intensity (0–1)")
plt.xticks(rotation=20)
plt.show()
```

Facetgrid

```python
g = sns.FacetGrid(df, col="EducationLevel", hue="Diagnosis", palette="Set1", height=4)
g.map(sns.scatterplot, "Age", "MMSE", alpha=0.5)
g.add_legend()
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle("Impact of Education on Age-Related Cognitive Decline")
plt.show()
```
